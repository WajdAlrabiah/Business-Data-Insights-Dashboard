```python
import pandas as pd
import numpy as np

df = pd.read_csv("bussiness.csv")
print("Shape BEFORE cleaning:", df.shape)
df.head()

```

```python
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(r"\s+", "_", regex=True)
)
df.columns.tolist()
```

```python
df.dtypes, df.isna().sum().sort_values(ascending=False).head(10)
```

```python
for col in df.columns:
    df[col] = df[col].astype("object")


df["market_cap"] = (
    df["market_cap"].astype(str)
        .str.replace(",", "", regex=False)
        .str.replace("$", "", regex=False)
        .str.replace("M", "e6", regex=False)
        .str.replace("B", "e9", regex=False)
        .str.replace("SAR", "", regex=False)
        .str.strip()
)
df["market_cap"] = pd.to_numeric(df["market_cap"], errors="coerce").astype("float")

df["employees"] = (
    df["employees"].astype(str)
        .str.replace(",", "", regex=False)
        .str.strip()
)
df["employees"] = pd.to_numeric(df["employees"], errors="coerce").astype("float")

print(df.dtypes)
```

```python
df.isna().sum()
```

```python
# Fill numeric NaN with median
df["market_cap"] = df["market_cap"].fillna(df["market_cap"].median())
df["employees"] = df["employees"].fillna(df["employees"].median())

# Fill categorical NaN with "Unknown"
text_cols = ["ticker", "name", "sector", "industry", "country", "currency", "description"]
for col in text_cols:
    df[col] = df[col].fillna("Unknown")
```

```python
print("Duplicates before:", df.duplicated().sum())
df = df.drop_duplicates()
print("Duplicates after:", df.duplicated().sum())
```

```python
def cap_outliers_iqr(s: pd.Series, k: float = 1.5):
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    low, high = q1 - k*iqr, q3 + k*iqr
    return s.clip(lower=low, upper=high)

df["market_cap"] = cap_outliers_iqr(df["market_cap"])
df["employees"] = cap_outliers_iqr(df["employees"])
```

```python
import numpy as np
df["log_market_cap"] = np.log1p(df["market_cap"])
df["log_employees"] = np.log1p(df["employees"])
```

```python
print("Duplicates:", df.duplicated().sum())
print("Missing:", df.isna().sum().sum())
print(df.dtypes)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Raw histograms
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(df["market_cap"], bins=30, ax=axes[0], kde=True)
axes[0].set_title("Histogram - Market Cap")

sns.histplot(df["employees"], bins=30, ax=axes[1], kde=True)
axes[1].set_title("Histogram - Employees")

plt.show()
```
